```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme()

hs = [0.2382, 0.1191, 0.0595, 0.0297, 0.01488, 0.00744, 0.00372]

L2s = [0.0576, 0.0105, 0.002219, 0.0004863, 0.0001103, 2.56744e-5, 6.00303e-6]
H1s = [0.9017, 0.3792, 0.17501, 0.084445, 0.0415808, 0.020654, 0.010305]

plt.loglog(hs, L2s, label=r"$L_2$ error")
plt.loglog(hs, H1s, label=r"$H_1$ error")
plt.loglog(hs, hs, label=r"$O(h)$")
plt.loglog(hs, np.array(hs) ** 2, label=r"$O(h^2)$")

plt.legend(loc="best")
```
